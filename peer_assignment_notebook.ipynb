```
data = {
    "Name": ["Alex", "Sam", "John", "Riya"],
    "Age": [25, 21, 29, 23],
    "Score": [85, 90, 76, 88]
}

df = pd.DataFrame(data)
df
```

```
df.describe()
```

```
plt.figure(figsize=(6,4))
plt.plot(df["Name"], df["Score"], marker="o")
plt.title("Student Scores")
plt.xlabel("Name")
plt.ylabel("Score")
plt.grid()
plt.show()
```
